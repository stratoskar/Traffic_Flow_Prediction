### Plot the road network of San Francisco, California
This is the background road netwrok, which is used extensively in our research.

```python
# import libraries
import osmnx as ox
import matplotlib.pyplot as  plt
```

```python
# specify the city name
city_name = "San Francisco, California, USA"
```

```python
# retrieve the driving street network
G = ox.graph_from_place(city_name, network_type="drive")
```

```python
# create a larg plot graph with the road network
fig, ax = ox.plot_graph(ox.project_graph(G), figsize=(15, 15), bgcolor='k', node_size=0, edge_color='w', edge_linewidth=0.5)
```
